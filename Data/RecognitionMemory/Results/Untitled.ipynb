```python
import os
import shutil
import re
from datetime import datetime

# ---- Configuration ----
input_dir = "/Users/bjm/Documents/School/MIT/labs/mcdermott/Tsimane2025/Data/RecognitionMemory/Results"  # 🔁 update this
backup_dir = os.path.join(input_dir, "backup_originals")
log_file = os.path.join(input_dir, "texture_renamed_ids.txt")

# ---- Create backup directory if it doesn't exist ----
os.makedirs(backup_dir, exist_ok=True)

# ---- Define date range ----
valid_dates = {f"{day}-Jul-2025" for day in range(24, 28)}  # 24 to 27 July

# ---- Regex pattern to extract relevant parts ----
pattern = re.compile(r"^(\d{4}-[A-Z]{3}-[A-Z]{3}_[^_]+)_(.+)_(\d{1,2}-Jul-2025)_Globalized-Music\.mat$")

renamed_ids = []

# ---- Process files ----
for filename in os.listdir(input_dir):
    if filename.endswith("Globalized-Music.mat"):
        match = pattern.match(filename)
        if match:
            subject_id, middle_part, date_str = match.groups()

            if date_str in valid_dates:
                old_path = os.path.join(input_dir, filename)
                new_filename = f"{subject_id}_{middle_part}_{date_str}_Textures.mat"
                new_path = os.path.join(input_dir, new_filename)

                # Move original file to backup
                shutil.copy2(old_path, os.path.join(backup_dir, filename))

                # Rename file in place
                os.rename(old_path, new_path)

                # Record subject ID
                renamed_ids.append(subject_id)

# ---- Save subject IDs to log file ----
with open(log_file, "w") as f:
    for subject_id in sorted(set(renamed_ids)):
        f.write(subject_id + "\n")

print(f"Renamed {len(renamed_ids)} files with Globalized-Music from July 24–27. Backup saved in '{backup_dir}'. Log written to '{log_file}'.")
```
