## Tiny power simulation

Under H0 (all three true correlations equal), the recentered-null rejection rate should be ~0.05 at $\alpha=0.05$. With only 30 reps this is noisy — you should see something in [0, 0.20]. The straddle-zero rate is typically much lower at small N — that's the small-N pathology in action.

```python
res = power_simulation_paired_bootstrap(
    n_reps=30, n_a=20, n_b=20, n_c=20, n_items=40,
    rho=(0.4, 0.4, 0.4),  # H0: all equal
    n_boot=500, seed=42, progress_every=0,
)
```

```python
print('reject_null     :', res.reject_null)
print('reject_straddle :', res.reject_straddle)
print('CI coverage     :', res.coverage)
print('mean obs r      :', res.mean_obs_r)
print('true differences:', res.true_diff)
```
